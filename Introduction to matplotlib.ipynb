```python
car_sales["Price"]=car_sales["Price"]/100
```

```python
car_sales
```

```python
car_sales["Sales_date"]=pd.date_range("1/1/2023",periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales["Total sales"]=car_sales["Price"].cumsum()
```

```python
car_sales
```

```python
heart_disease=pd.read_csv("heart-disease.csv")
```

```python
heart_disease.head()
```

```python
plt.hist(heart_disease["age"],bins=113);
```

```python
heart_disease.plot(figsize=(100,10),subplots=True)
```

```python
over_50=heart_disease[heart_disease["age"]>50]
```

```python
over_50.plot(kind="scatter",x='age',y='chol')
```

```python
# pyplot method
over_50.plot(kind="scatter",x='age',y='chol',c='target')
```

```python
# oo method mixed with pyplot method
fig,ax=plt.subplots(figsize=(10,6));
over_50.plot(kind="scatter",x='age',y='chol',ax=ax,c='target')
# ax.set_xlim([45,100])
```

```python
# complete oo method from scratch
fig, ax= plt.subplots(figsize=(10,6))
scgraph=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
ax.set(xlabel="age",ylabel="chol",title="shitty ass graph")
ax.legend(*scgraph.legend_elements(),title="target")
ax.axhline(over_50["chol"].mean(),linestyle="dotted")
```

```python
fig, (ax0,ax1)= plt.subplots(ncols=1,nrows=2,figsize=(10,10),sharex=True)
graph1=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax0.set(ylabel="cholestrol",title="heart disease graph wrt cholestrol");
ax0.legend(*graph1.legend_elements(),title="legend1")
ax0.axhline(over_50["chol"].mean(),linestyle="--")

graph2=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.set(xlabel="age",ylabel="maxheartrate",title="heart disease graph wrt maxheartrate")
ax1.legend(*graph2.legend_elements(),title="legend2")
ax1.axhline(over_50["thalach"].mean(),linestyle="dotted")

fig.suptitle("welcome world",fontsize=15,fontweight="bold")
```

```python
plt.style.available
plt.style.use("seaborn-v0_8-whitegrid")
```

```python
plt.style.use("seaborn-v0_8-darkgrid")
fig, (ax0,ax1)= plt.subplots(ncols=1,nrows=2,figsize=(10,10),sharex=True)
graph1=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
ax0.set(ylabel="cholestrol",title="heart disease graph wrt cholestrol");
ax0.legend(*graph1.legend_elements(),title="legend1")
ax0.axhline(over_50["chol"].mean(),linestyle="--")
graph2=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="cividis")
ax1.set(xlabel="age",ylabel="maxheartrate",title="heart disease graph wrt maxheartrate")
ax1.legend(*graph2.legend_elements(),title="legend2")
ax1.axhline(over_50["thalach"].mean(),linestyle="dotted")

fig.suptitle("welcome world",fontsize=15,fontweight="bold")
```
